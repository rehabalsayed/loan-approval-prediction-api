```python
import pandas as pd

df = pd.read_csv("../data/loan_approval_dataset.csv")

df.columns = df.columns.str.strip()
```

```python
df = df.drop(columns=['loan_id'])
```

# Negative Values

```python
asset_columns = [
    'residential_assets_value',
    'commercial_assets_value',
    'luxury_assets_value',
    'bank_asset_value'
]

for col in asset_columns:
    print(col)
    print((df[col] < 0).sum())
```

```python
for col in asset_columns:

    negative_count = (df[col] < 0).sum()

    if negative_count > 0:

        median_val = df.loc[
            df[col] >= 0,
            col
        ].median()

        df[col] = df[col].apply(
            lambda x: median_val if x < 0 else x
        )

        print(f"{col}: {negative_count} negatives replaced")
```

```python
for col in asset_columns:
    print(col, (df[col] < 0).sum())
```

# encoding

```python
print(df['education'].value_counts())

print(df['self_employed'].value_counts())

print(df['loan_status'].value_counts())
```

```python
df['education'] = (
    df['education']
    .str.strip()
    .map({
        'Graduate': 1,
        'Not Graduate': 0
    })
)

df['self_employed'] = (
    df['self_employed']
    .str.strip()
    .map({
        'Yes': 1,
        'No': 0
    })
)

df['loan_status'] = (
    df['loan_status']
    .str.strip()
    .map({
        'Approved': 1,
        'Rejected': 0
    })
)
```

# Feature Engineering

```python
df['total_assets'] = (
    df['residential_assets_value']
    + df['commercial_assets_value']
    + df['luxury_assets_value']
    + df['bank_asset_value']
)
```

```python
df['loan_to_income_ratio'] = (
    df['loan_amount']
    / (df['income_annum'] + 1)
)
```

```python
df['annual_payment'] = (
    df['loan_amount']
    / df['loan_term']
)
```

```python
df['debt_service_ratio'] = (
    df['annual_payment']
    / (df['income_annum'] + 1)
)
```

```python
df['assets_to_loan_ratio'] = (
    df['total_assets']
    / (df['loan_amount'] + 1)
)
```

```python
df['income_per_dependent'] = (
    df['income_annum']
    / (df['no_of_dependents'] + 1)
)
```
